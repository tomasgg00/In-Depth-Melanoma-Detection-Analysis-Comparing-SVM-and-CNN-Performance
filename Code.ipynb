# Loading the data - Resizing will be done after the filtering

```python
# Loading of the images and creation of the Dataframe
def load_images(base_path, categories):
    data = {'photo_id': [], 'image': [], 'image_array': [], 'category': [], 'subset': []}
    
    for subset in ['train', 'test']:
        for category in categories.keys():
            folder_path = os.path.join(base_path, subset, category)
            for filename in os.listdir(folder_path):
                if filename.lower().endswith(('.jpg', '.jpeg')):  # Check for both .jpg and .jpeg extensions
                    file_path = os.path.join(folder_path, filename)
                    # Load the image
                    image = load_img(file_path)
                    # Add data to the DataFrame
                    data['photo_id'].append(filename)
                    data['image'].append(image)
                    data['category'].append(categories[category])
                    data['subset'].append(subset)
                    data['image_array'].append(None) 
    
    df = pd.DataFrame(data)
    return df

# Load the DataFrame
categories = {'Benign': 0, 'Malignant': 1, 'Undetected': 2}
base_path = '/work/Skin-Cancer-Detection-A-Comparative-Analysis-of-SVM-and-CNN-on-Melanoma-Image-Classification/images'
df = load_images(base_path, categories)
```

```python
# Compute three different types of hashes: Average Hash (aHash), Difference Hash (dHash), and Wavelet Hash (wHash).
def compute_hashes(df):
    a_hashes = []
    d_hashes = []
    w_hashes = []
    #
    for image in df['image']:
        # Compute the Average Hash (aHash)
        a_hashes.append(str(imagehash.average_hash(image)))
        # Compute the Difference Hash (dHash)
        d_hashes.append(str(imagehash.dhash(image)))
        # Compute the Wavelet Hash (wHash)
        w_hashes.append(str(imagehash.whash(image)))
    df['a_hash'] = a_hashes
    df['d_hash'] = d_hashes
    df['w_hash'] = w_hashes

compute_hashes(df)

# Check for duplicates based on different hashes
duplicate_a_hash = df[df.duplicated(subset='a_hash', keep=False)]
duplicate_d_hash = df[df.duplicated(subset='d_hash', keep=False)]
duplicate_w_hash = df[df.duplicated(subset='w_hash', keep=False)]

# Print all the duplicates detected from each method
print("Duplicate Images based on Average Hash:")
print(duplicate_a_hash[['photo_id', 'category', 'subset', 'a_hash']])

print("Duplicate Images based on Difference Hash:")
print(duplicate_d_hash[['photo_id', 'category', 'subset', 'd_hash']])

print("Duplicate Images based on Wavelet Hash:")
print(duplicate_w_hash[['photo_id', 'category', 'subset', 'w_hash']])
```

```python
# Function to visualize duplicate images
def visualize_duplicates(df, hash_column):
    duplicates = df[df.duplicated(subset=hash_column, keep=False)]
    # Group the duplicates by the hash values
    grouped = duplicates.groupby(hash_column)
    for hash_value, group in grouped:
        # Check if there is more than one duplicate for the given hash value
        if len(group) > 1:
            # Create a subplot with the number of columns equal to the number of duplicates
            fig, axes = plt.subplots(1, len(group), figsize=(15, 5))
            # Plot the duplicates
            for ax, (_, row) in zip(axes, group.iterrows()):
                ax.imshow(row['image'])
                ax.set_title(f"{row['photo_id']}\n{row['category']}\n{row['subset']}")
                ax.axis('off')
            plt.suptitle(f"Duplicates based on {hash_column}: {hash_value}")
            plt.show()

# Visualize duplicates for each hash type
#Only the first five because of length reasons, if all printed impossible to download, but we examined all of them
visualize_duplicates(df[df['a_hash'].duplicated()].head(5), 'a_hash')
visualize_duplicates(df[df['d_hash'].duplicated()].head(5), 'd_hash')
visualize_duplicates(df[df['w_hash'].duplicated()].head(5), 'w_hash')
```

Check the size of the images before resize them

```python
# Function to calculate the aspect ratio of an image
def aspect_ratio(image):
    width, height = image.size
    return width / height

# Function to filter images based on aspect ratio
def filter_aspect_ratio(df_cleaned):
    filtered_df = df_cleaned[df_cleaned.apply(lambda row: 0.25 <= aspect_ratio(row['image']) <= 4.0, axis=1)]
    return filtered_df

# Apply the aspect ratio filter to the DataFrame
df_aspect_filtered = filter_aspect_ratio(df_cleaned)

# Print the size of the Dataframes
print(f"Original DataFrame size: {len(df_cleaned)}")
print(f"Aspect Ratio Filtered DataFrame size: {len(df_aspect_filtered)}")
```

Check the average RGB values

```python
# Function to calculate the average RGB values of an image
def average_rgb(image):
    np_image = np.array(image)
    avg_r = np.mean(np_image[:, :, 0])
    avg_g = np.mean(np_image[:, :, 1])
    avg_b = np.mean(np_image[:, :, 2])
    return avg_r, avg_g, avg_b

# Function to filter images based on average RGB values and return filtered out images
def filter_rgb_values(df):
    filtered_out_indices = []
    for idx, row in df.iterrows():
        # Calculate the average RGB values for the image in the row
        avg_r, avg_g, avg_b = average_rgb(row['image'])
        # Check if the average RGB values are within the range (5 to 250)
        if not (5 <= avg_r <= 250 and 5 <= avg_g <= 250 and 5 <= avg_b <= 250):
            # If the images are out of range add the row index to the filtered_out_indices list
            filtered_out_indices.append(idx)
    # Create a DataFrame containing only the rows with filtered out indices
    filtered_out_df = df.loc[filtered_out_indices]
    # Create a DataFrame containing the rows that were not filtered out
    filtered_df = df.drop(filtered_out_indices)
    # Return the two Dataframes created
    return filtered_df, filtered_out_df

# Apply the RGB value filter to the aspect ratio filtered DataFrame
df_final_filtered, df_filtered_out = filter_rgb_values(df_aspect_filtered)

# Print the size of the Dataframes
print(f"Aspect Ratio Filtered DataFrame size: {len(df_aspect_filtered)}")
print(f"Final Filtered DataFrame size: {len(df_final_filtered)}")
print(f"Filtered Out DataFrame size: {len(df_filtered_out)}")

# Function to visualize filtered out images
def visualize_filtered_images(df):
    for idx, row in df.iterrows():
        image = row['image']
        plt.imshow(image)
        plt.title(f"Photo ID: {row['photo_id']}\nCategory: {row['category']}\nSubset: {row['subset']}")
        plt.axis('off')
        plt.show()

# Visualize the filtered out images
# Printed only 5 because otherwise impossible to download file
visualize_filtered_images(df_filtered_out.head(5))
```

# CNN - Preparing the dataset

```python
### Load the dataset ###
 
# One-hot encode the 'category' column into necessary format for training a neural network in TensorFlow
df['category_encoded'] = df['category'].apply(lambda x: to_categorical(x, num_classes=3))

# Splitting data into training, validation and test sets
test_df = df[df['subset'] == 'test']
train_df, val_df = train_test_split(df[df['subset'] == 'train'], test_size=0.20, random_state=42)  # This will make 80% train, 20% validation

# Checking datasets dimension
print(len(test_df))
print(len(train_df))
print(len(val_df))
```

```python
#   Function to convert a DataFrame to a TensorFlow dataset to optimise perfomance
def df_to_dataset(dataframe, shuffle=True, batch_size=32, augment=False):
    # Stack image arrays and one-hot encoded labels from the DataFrame
    images = np.stack(dataframe['image_array'].values)
    labels = np.stack(dataframe['category_encoded'].values)
    # Create a TensorFlow dataset from the stacked images and labels
    dataset = tf.data.Dataset.from_tensor_slices((images, labels))
    
     # Shuffle the dataset to randomize the order of the data
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(dataframe))
    
    # Batch the dataset and prefetch
    dataset = dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset
```

```python
# Apply the function -> Create TensorFlow datasets for training, validation, and testing
train_ds = df_to_dataset(train_df, shuffle=True, batch_size=32)
val_ds = df_to_dataset(val_df, batch_size=32, shuffle=False)
test_ds = df_to_dataset(test_df, batch_size=32, shuffle=False)
```

# CNN

```python
# Define the architecture of the CNN model
model = Sequential([
     # First convolutional layer: 32 filters, 3x3 kernel size, ReLU activation
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),# Pooling layer to reduce dimensions
    BatchNormalization(),  # Normalize to speed up training

    # Second convolutional layer: 64 filters, 3x3 kernel size, ReLU activation
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2), # Pooling layer
    BatchNormalization(),  # Normalize again
    
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2), # pooling layer
    BatchNormalization(),  # Normalize again

    Flatten(), # Flatten the feature maps to 1D for the dense layer
    Dense(512, activation='relu'), # Used lectures coding as guidance
    Dropout(0.7),  # Dropout to avoid overfitting
    Dense(3, activation='softmax') # Output layer with softmax for multi-class classification (3 labels)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())

# Implement early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
```

```python
# Start timing
start_time_overall = time.time()

# Train the model
history = model.fit(
    train_ds,
    epochs=30,
    validation_data=val_ds
)
# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(test_ds)
print("Test Accuracy:", test_accuracy)

# Calculate and print the running time
overall_time_cnn1 = time.time() - start_time_overall
print(f"Total running time: {overall_time_cnn1:.2f} seconds.")
```
